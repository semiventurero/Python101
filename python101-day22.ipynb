```python
import numpy as np # numerical python
import pandas as pd # data processing
from pandas import read_excel
import pylab as py 
import plotly as plt3
plt3.offline.init_notebook_mode(connected=True)
import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot, plot
import plotly.io as pio
from pandas_datareader import data as web

import warnings#ignore warnings
warnings.filterwarnings("ignore")


print("Pandas version:",pd.__version__)
print("Numpy version:",np.__version__)
print("Plotly version:",plt3.__version__)
```

```python
df2 = pd.read_csv("data/currency.csv")
df2.head()
```

```python
stock = 'AAPL'
stock2 = "NDAQ" 
df = web.DataReader(stock, data_source='yahoo', start='01-01-2021')
df2 = web.DataReader(stock2, data_source='yahoo', start='01-01-2021')
```

```python
trace1 = {
    'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low,
    'type': 'candlestick',
    'name': 'AAPL',
    'showlegend': True
}
trace4 = {
    'x': df2.index,
    'open': df2.Open,
    'close': df2.Close,
    'high': df2.High,
    'low': df2.Low,
    'type': 'candlestick',
    'name': 'NDAQ',
    'showlegend': True
}
```

```python
import plotly.graph_objects as go
font1 = {'family':'serif','color':'black','size':20}
data = [trace1]
# Config graph layout
layout = go.Layout({
    'title': {
        'text': 'Apple(AAPL) Moving Averages -YTD',
        'font': font1
    }
},paper_bgcolor='rgba(0,0,0)',
    plot_bgcolor='rgba(94,102,115)')#background color
fig = go.Figure(data=data, layout=layout)
fig.update_layout(
    title='Apple(AAPL) Moving Averages -YTD',
    xaxis_title="Date",
    yaxis_title="Prices",
    font=dict(
        family="serif",
        size=18,
        color="white"
    )
)
fig.show()
pio.write_html(fig, file='htmls/AAPL1_blackbackground.html', auto_open=True)
#plotly.offline.plot(fig, filename="htmls/AAPL1_blackbackground.html")
```

```python
import plotly.graph_objects as go
font1 = {'family':'serif','color':'black','size':20}
data = [trace1,trace4]
# Config graph layout
layout = go.Layout({
    'title': {
        'text': 'Apple(AAPL) Moving Averages -YTD',
        'font': font1
    }
},paper_bgcolor='rgba(0,0,0)',
    plot_bgcolor='rgba(94,102,115)')#background color
fig = go.Figure(data=data, layout=layout)
fig.update_layout(
    title='NASDAQ and APPLE',
    xaxis_title="Date",
    yaxis_title="Prices",
    font=dict(
        family="serif",
        size=18,
        color="white"
    )
)
fig.show()
pio.write_html(fig, file='htmls/nasdaq&apple.html', auto_open=True)
```

```python
# Calculate and define moving average of 30 periods
avg_30 = df.Close.rolling(window=20, min_periods=1).mean()

# Calculate and define moving average of 50 periods
avg_50 = df.Close.rolling(window=40, min_periods=1).mean()

trace2 = {
    'x': df.index,
    'y': avg_30,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 2,
        'color': 'blue'
            },
    'name': 'Moving Average of 30 periods'
}

trace3 = {
    'x': df.index,
    'y': avg_50,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'red'
    },
    'name': 'Moving Average of 50 periods'
}
data2 = [trace1,trace2, trace3]
fig = go.Figure(data=data2, layout=layout)
fig.update_layout(
    title='Apple(AAPL) Moving Averages -YTD',
    xaxis_title="Date",
    yaxis_title="Prices",
    font=dict(
        family="serif",
        size=18,
        color="white"
    )
)
fig.show()
pio.write_html(fig, file='htmls/apple_movingavs.html', auto_open=True)
```
