# Multiple Plots

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
import numpy as np
import plotly

fig = make_subplots(
    rows=2, cols=1,
    )#row_heights=[1., 1., 1.]
fig.add_trace(go.Scatter(
                    x = train.Age,
                    y = train.Fare,
                    mode = "lines+markers",
                    name = "fare/age",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= train.Name,
                    ),
             row=1, col=1
             )
fig.add_trace(
    go.Scatter(x = train.Age,
                    y = train.Fare,
                    mode = "markers",
                    name = "fare/age",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= train.Name
    ),
    row=2, col=1
)
plotly.offline.plot(fig, filename="htmls/fare-age3.html")
```

```python
df = train.copy()
df = df.drop(["Pclass","Name","Age","SibSp","Parch","Ticket","Fare","Cabin","Embarked"], axis=1)

df1 = df[df["Survived"] == 0]
df2 = df[df["Survived"] == 1]
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

fig = make_subplots(
    rows=5, cols=1,
    )#row_heights=[1., 1., 1.]
fig.add_trace(go.Scatter(
                    x = train.Age,
                    y = train.Fare,
                    mode = "lines+markers",
                    name = "fare/age",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= train.Name,
                    ),
             row=1, col=1
             )
fig.add_trace(
    go.Scatter(x = train.Age,
               y = train.Fare,
               mode = "markers",            
               name = "fare/age",
               marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
               text= train.Name
    ),
    row=2, col=1
)
fig.add_trace(
    go.Box(
    x=train[train["Sex"]=="male"].Age,
    #y=dataframe.teaching,
    name = "values",
    marker = dict(color = "red"),
  ),
    row=3, col=1
)
fig.update_layout(
    title = "Multiple Plots",
     titlefont = font1
)

fig.add_trace(
    go.Bar(x=df["Sex"],
    y=df2.value_counts(),
    name="Survived",
    marker_color='green'
        
    ),
    row=4, col=1
)
fig.add_trace(go.Bar(
    x=df["Sex"],
    y=df1.value_counts(),
    name= "Died",
    marker_color="red"),
             row=5, col=1
             )

plotly.offline.plot(fig, filename="htmls/fare-age3.html")

```

```python
import plotly.graph_objects as px


plot = px.Figure(data=[px.Scatter(
    x = train.Age,
    y = train.Fare,
    mode='markers',
    name = "fare/age",
    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
    text= train.Name)
])
plot.update_layout(title = 'Fare/Age',
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "bar"],
                    label="Bar Chart",
                    method="restyle"
                )
            ]),
            direction="down",
        ),
    ]
)

plot.show()

#pio.write_html(fig2, file='htmls/fare-age2.html', auto_open=True)
```
