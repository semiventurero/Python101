seek(), write, r+(to read and write at the same time)

```python
with open("file3.txt","r+",encoding = "utf-8") as file: 
    file.seek(20) # 20. byte
    file.write("5353435. try")
    file.seek(50) # 50. byte
    file.write("5353436. try\n")
    file.seek(20000) #if you write a byte which is larger than the bytes in file. It will write in the end of the file
    file.write("53534359898 try")
```

```python
with open("file3.txt","r+",encoding = "utf-8") as file:
    print(file.read())
```

# changing in the begining of the file

```python
with open("file3.txt","r+",encoding = "utf-8") as file:
    content = file.read()
    print(content)
```

```python
with open("file3.txt","r+",encoding = "utf-8") as file:
    content = file.read()
    
    content = "END\n" + content
    file.seek(0)
    file.write(content)
```

```python
with open("file3.txt","r+",encoding = "utf-8") as file:
    content = file.read()
    print(content)
```

# Delete

```python
#Code source: https://stackoverflow.com/questions/4710067/how-to-delete-a-specific-line-in-a-file
with open("file3.txt", "r") as f:
    lines = f.readlines()
with open("file3.txt", "w") as f:
    for line in lines:
        if line.strip("\n") != "END":
            f.write(line)
```

```python
with open("file3.txt","r+",encoding = "utf-8") as file:
    content = file.read()
    print(content)
```
